Handle The Inconsistencies in the Data

```python
import pandas as pd
```

```python
#Inconsistent Data
#create a sample dataframe with inconsistencies
data = {'Name': ['Alice','Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 25, -30, 35, 40, 28],
        'Salary': [50000, 50000, 60000, -70000, 80000, 90000],
        'Department': ['HR', 'HR', 'Finance', 'IT', 'IT', 'Finance'],
        'country': ['USA', 'USA', 'USA', 'America', 'United States', 'USA']
        }
df = pd.DataFrame(data)
df
```

```python
#Handle Inconsistencies
df['country'] = df['country'].replace({'America': 'USA', 'United States': 'USA'})
df['Age'] = df['Age'].abs()
df['Salary'] = df['Salary'].abs()
df.drop_duplicates(inplace=True)
df

#And that's how you handle inconsistencies in the data! obviously, the methods may vary based on the specific inconsistencies present in your dataset.
```
